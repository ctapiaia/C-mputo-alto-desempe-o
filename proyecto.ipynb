```python
import skimage.io as io

def t_fourier(imag,salida):
    print(imag)
    texto='I:/datas/proyecto/COVID'+'\\' +imag
    covid = cv2.imread(texto,0)
    #print(covid)
    #ms = io.imread(img)
    #print(ms)
    #print(ms.shape,ms.shape[2])
    # las imgenes originales tienen 13 bandas
    #assert ms.shape[2] == 13
    dft = cv2.dft(np.float32(covid), flags=cv2.DFT_COMPLEX_OUTPUT)
    dft_shift = np.fft.fftshift(dft)

    magnitude_spectrum = 20 * np.log(cv2.magnitude(dft_shift[:, :, 0], dft_shift[:, :, 1]))
    rows, cols = covid.shape
     
    crow, ccol = int(rows / 2), int(cols / 2)

    mask = np.ones((rows, cols, 2), np.uint8)

    r = 1.1

    center = [crow, ccol]
    x, y = np.ogrid[:rows, :cols]
    mask_area = (x - center[0]) ** 2 + (y - center[1]) ** 2 <= r*r
    mask[mask_area] = 0
    
    fshift = dft_shift * mask

    fshift_mask_mag = 2000 * np.log(cv2.magnitude(fshift[:, :, 0], fshift[:, :, 1]))
    
    f_ishift = np.fft.ifftshift(fshift)
    img_back = cv2.idft(f_ishift)
    img_back = cv2.magnitude(img_back[:, :, 0], img_back[:, :, 1])
    
    fig = plt.figure(figsize=(12, 12))
    ax1 = fig.add_subplot(2,2,1)
    ax1.imshow(covid, cmap='gray')
    ax1.title.set_text('Input Image')
    
    ax2 = fig.add_subplot(2,2,2)
    ax2.imshow(magnitude_spectrum, cmap='gray')
    ax2.title.set_text('FFT of image')
    
    ax3 = fig.add_subplot(2,2,3)
    ax3.imshow(fshift_mask_mag, cmap='gray')
    ax3.title.set_text('FFT + Mask')
    
    ax4 = fig.add_subplot(2,2,4)
    ax4.imshow(img_back, cmap='gray')
    ax4.title.set_text('After inverse FFT')
    # plt.show()
    file_path=salida+'sal_normal_'+imag
    print(file_path)
    #plt.savefig(file_path)
    os.chdir(salida)
    #cv2.imwrite(imag, fig)
    plt.savefig(imag)
    #io.imsave(file_path, fig)
    plt.close(fig)
 
    return covid

#dest_folder = f'{path}/results'
def save_png(name, img, texto, sep="\\"):
    
    
    img_name = name.split(sep)[-1][:-4]
    file_p = texto+img_name+'.png'
    io.imsave(file_p, img)
    return file_path
```

```python
from concurrent.futures import ThreadPoolExecutor

start_time = time.perf_counter()
with ThreadPoolExecutor() as executor:
    executor.map(generate_transformada, contenido)

finish_time = time.perf_counter()
print(f"Program finished in {finish_time-start_time} seconds")    
```

```python
import multiprocessing
from concurrent.futures import ProcessPoolExecutor

start_time = time.perf_counter()

num_cores = multiprocessing.cpu_count()
with ThreadPoolExecutor(max_workers=num_cores) as pool:
    with tqdm(total=len(contenido)) as progress:
        futures = []
        for nombre in contenido:
            future = pool.submit(generate_transformada, nombre)
            future.add_done_callback(lambda p: progress.update())
            futures.append(future)

        for future in futures:
            future.result()
finish_time = time.perf_counter()
print(f"Program finished in {finish_time-start_time} seconds")    
```
